### 7. Metoda Monte Carlo

```python
import random
import numpy
from matplotlib import pyplot
```

```python
def narozeniny(pocet_iteraci: int, pocet_lidi: int) -> int:
    shoda_narozenin = 0
    pravdepodobnosti = []
    for iterace in range(1, pocet_iteraci + 1):
        # pomocí komprehenze generujeme n náhodných čísel, reprezentující dny v roce
        bdays = {random.randint(1, 365) for _ in range(pocet_lidi)}
        # nerovnost znamená ze v množině byli generovány duplicity a tudíž shoda narozenin
        if len(bdays) != pocet_lidi:
            shoda_narozenin += 1
        pravdepodobnosti.append(shoda_narozenin/iterace)
    return pravdepodobnosti
```

```python
y = []
for pocet_lidi in range(60):
    x = narozeniny(10000, pocet_lidi)
    y.append(x[-1])

# hledání indexu (počtu lidí), která se nejvíc blíží 0.50
z = [abs(i - 0.50) for i in y]
print(z.index(min(z)))

pyplot.plot(y)
pyplot.xlabel("počet lidí")
pyplot.ylabel("pravděpodobnost shody narozenin")
pyplot.axhline(0.5, color="r", linestyle=":")
pyplot.grid()
pyplot.show()
```

```python
pocet_lidi = 23
y = narozeniny(100000, pocet_lidi)
print(f"Pravděpodobnost, že alespoň 2 lidi, ve skupine {pocet_lidi} lidí, mají stejné narozeniny: {y[-1] * 100} %")

pyplot.plot(y, label="naše data")
pyplot.ylim(0.3, 0.7)
pyplot.xlabel("počet iterací")
pyplot.ylabel("pravděpodobnost shody narozenin")
pyplot.axhline(0.507, color="r", linestyle=":", label="tvrzená hodnota")
pyplot.legend()
pyplot.show()
```
